```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/krpadilla8/BCIS321/refs/heads/main/top_100_books-2020-Katelyn_Padilla.csv')
print(df)
```

```python
df.drop(["Unnamed: 0"], axis=1, inplace=True)
df.drop(["Unnamed: 4"], axis=1, inplace=True)
df.drop(["Unnamed: 9"], axis=1, inplace=True)

#df.loc[86]=[88, "Man's Search for Meaning", "Viktor E. Frankl", 4.7, 4959, "Paperback", 0]
df.loc[86,6] = 9.99
df.head()
```

```python
#Paperback and Hardcover
changes = 0
i = 0
for t in df.type:
  if t == "Board Book":
    df.type.replace("Board Book", "Hardcover", inplace=True)
    changes +=1
  if t == "Mass Market Paperback":
    df.type.replace("Mass Market Paperback", "Paperback", inplace=True)
    changes +=1
  i +=1
print(changes)
df.head()
```

```python
#Rating by Price
rating = df["rating"]
price = df["price"]
rating_by_price = rating/price
print(rating_by_price)
df.insert(4, "rating_by_price", rating_by_price)
df.sort_values(by=["rating_by_price"], ascending=False)
df.head()
```

```python
#Updated CSV file
df.to_csv('Katelyn_Padilla_top_100_books_cleaned.csv')
```
